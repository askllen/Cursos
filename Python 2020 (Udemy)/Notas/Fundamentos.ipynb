# Operadores Aritiméticos

```python
# Operadores aritiméticos são binários, que precisa de dois operandos para operar, como abaixo.
# Nomeclatura: Infix, ela fica entre os dois operandos

print(2 + 3)
print(4 - 7)
print(2 * 5.3)
print(9.4 / 3)

print(9.4 // 3) #Resultado obrigatoriamente inteiro.

print(2 ** 8) #Exponenciação ** eleva o numero a potencia
print(10 % 3) #Modulo é o resto da divisão

a = 12
b = a
print(a + b)
```

```python
# Desafio: Qual o percentual das despesas em relação ao salário?

salario = 3450.45
despesas = 2456.2

print(salario % despesas)

# Resposta do desafio
percentual_comprometido = despesas / salario * 100
percentual_comprometido
```

# Operadores Relacionais

```python
print(3 > 4)
print(4 >= 3)
print(1 < 2)
print(3 <= 1)
print(3 != 2) #Diferente
print(3 == 3) #Notar diferença entre atribuição e comparador e o sinal de igualdade
print(2 == '2') #O segundo item '2' é uma string por estar entre aspas, logo não é um valor
```

# Operadores de Atribuição

```python
a = 3
a = a + 7
print(a)

#a = 5 
a += 5 #Igual porem simplificado: a = a + 5
print(a)

a -= 3
print(a)

a *= 2
print(a)

a /= 4
print(a)

a %= 4 #Atribuição modular
print(a)

a **= 8 #Atribuição de exponenciação
print(a)

a //= 127
print(a)
```
